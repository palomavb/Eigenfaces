# Eigenfaces and face recognition

O conceito de eigenfaces consiste no conjunto de autovetores de uma matriz de covariância formada por imagens de faces (rostos). Sirovich e Kirby desenvolveram essa ideia para representar padrões encontrados nas imagens utilizando o método de PCA (Principal Component Analisys). Em 1991, Turk e Pentland desenvolveram um método para avaliar a distância entre rostos e “espaços de faces” para determinar quais rostos são similares, possibilitando, assim, o reconhecimento facial. 

O reconhecimento facial se baseia em identificar padrões faciais tais como formato da boca, rosto, distância entre olhos, entre outros. 
A técnica PCA é utilizada a fim de extrair os autovetores que são as principais componentes que representam da melhor maneira o cojunto de dados. Na deteccão de padrões, emprega-se geralmente a distância euclidiana como critério de classificação.

## Como o algoritmo funciona?

Se a partir de um conjunto de M imagens diferentes de rostos humanos, todas com a mesma dimensão (por exemplo, todas imagens em escala de cinza de dimensão $m×n$ pixels), vetorizarmos cada imagem em $L=m×n$ pixels, podemos apresentar todo o conjunto de dados como uma matriz $A=L×M$, onde cada elemento da matriz é o valor da escala de cinza do pixel.

Ao aplicar a técnica de PCA, o resultado são os vetores chamados componentes principais e cada um deles possui o mesmo comprimento que o comprimento da coluna da matriz  (por isso é importante que as imagens tenham o mesmo tamanho). Os diferentes componentes principais da mesma matriz são ortogonais uns aos outros, portanto, formam um espaço vetorial para o qual cada coluna da matriz pode ser representada como uma combinação linear dos componentes principais.

Primeiro tomamos $C = A - a$, onde $a$ é o vetor médio da matriz $A$. Assim, $C$ é a matriz que subtrai o vetor médio $a$ de cada coluna de $A$. 
Então, encontramos os autovalores e autovetores da matriz de covariância $S = C \cdot C^T$. Os componentes principais são estes autovetores na ordem decrescente dos autovalores. 

Os vetores componentes principais de $A$, ou autovetores de $S$, são as direções chave para construirmos as colunas da matriz $A$. A importância dos componentes principais pode ser determinada a partir dos autovalores correspondentes. Quanto maior for o autovalor, mais útil o componente principal, pois possui mais informação sobre $A$. Assim, podemos manter apenas os primeiros K vetores componentes principais. Se a matriz $A$ é o conjunto de dados de fotos de rostos, os primeiros K vetores componentes principais são as K "fotos de rosto" mais importantes, ou seja, a "eigenface picture".

Para qualquer foto de rosto dada, podemos projetar a sua versão com o vetor médio subtraído na eigenface picture utilizando produto vetorial. O resultado é o quão próxima esta imagem está relacionada com a eigenface. Se a foto estiver nenhum pouco relacionada com a eigenface, esperamos que o resultado seja zero. Para as K eigenfaces, podemos encontrar K produtos vetoriais para qualquer imagem de rosto. Podemos apresentar o resultado como "pesos" desta imagem de rosto em relação à eigenface. O peso é normalmente apresentado como um vetor.

Por outro lado, se tivermos um vetor de peso, podemos somar cada eigenface sujeita ao peso e reconstruir um novo rosto. Denotemos as eigenfaces como matriz F, que é uma matriz L×K, e o vetor de peso $w$ é um vector coluna. Depois, para qualquer $w$ podemos construir a imagem de um rosto como $z = F \cdot w$.

Como só estamos usando os K melhores vetores principais do componente, devemos esperar que a imagem resultante seja distorcida, mas mantendo alguma característica facial.

Uma vez que a matriz eigenface é constante para o conjunto de dados, um vetor de peso w que varia significa uma imagem facial variável. Por conseguinte, podemos esperar que as imagens da mesma pessoa forneçam vetores de peso semelhantes, mesmo que as imagens não sejam idênticas. Como resultado, podemos utilizar a distância entre dois vectores de peso como uma métrica de como duas figuras se assemelham.

## Implementando o código

```python
import cv2 #para converter as imagens em matrizes, segundo a tonalidade dos pixels
import zipfile
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

Utilizaremos o dataset disponível em https://www.kaggle.com/vinzard11/lfwgrey-crop. 

O ideal é que as imagens possuam o mesmo tamanho e que tenham os rostos alinhados no meio das imagens, pois isso facilita a remoção de ruídos. Para nossa sorte, as imagens do dataset que utilizaremos já foram formatadas. Utilizaremos cerca de 40 imagens.

Primeiramente, abrimos as imagens e as colocamos em um dicionário python


```python
faces = {}
with zipfile.ZipFile("Faces.zip", mode = 'r') as facezip:
    for file in facezip.namelist():
        with facezip.open(file) as image:
            faces[file] = cv2.imdecode(np.frombuffer(image.read(), np.uint8), cv2.IMREAD_GRAYSCALE)
print(faces.keys())
```

```python
faceshape = list(faces.values())[20].shape
print("Formato das imagens:", faceshape) #checando a quantidade de pixels das imagens
```

```python
print(list(faces.keys())[:5]) #checando o nome das imagens
```

```python
fig, axes = plt.subplots(4,5,sharex=True,sharey=True,figsize=(8,10))
faceimages = list(faces.values())[-20:] #apenas as matrizes das últimas 20 imagens
for i in range(20):
    axes[i%4][i//4].imshow(faceimages[i], cmap="gray") #a função imgshow plota as imagens a partir de suas matrizes
plt.show()
```

```python
classes = set(filename.split("_")[0] for filename in faces.keys()) #usamos o set para garantir que não existe valores duplicados
print("As classes são:", (classes))
print("Número de classes:", len(classes))
```

```python
facematrix = []
facelabel = []
for key,val in faces.items():
    if key.startswith("Faces/Alvaro_"):
        continue #tiramos todas as faces do Alvaro para comparar depois
    
    if key == "Faces/Abdullah_Gul_0009.pgm":
        continue #tiramos especificamente essa imagem para comparar depois
    
    facematrix.append(val.flatten()) #passa a array para uma dimensão
    facelabel.append(key.split("_")[0])  #nomes das imagens
 
#cria array com a quantidade de imagens (29) e o número de pixels de cada imagem
facematrix = np.array(facematrix)
print(facematrix.shape)
```

```python
#Usamos o PCA do sklearn
from sklearn.decomposition import PCA
pca = PCA().fit(facematrix)
print(pca.explained_variance_ratio_) #variancia para cada um dos componentes da matriz, ou seja, cada uma das imagens
```

```python

#Usamos todos os componentes do PCA para as imagens. Esse é o K
n_components = 29
eigenfaces = pca.components_[:n_components]
 
#Plotar as eigenfaces
fig, axes = plt.subplots(4,5,sharex=True,sharey=True,figsize=(8,10))
for i in range(20):
    axes[i%4][i//4].imshow(eigenfaces[i].reshape(faceshape), cmap="gray")
plt.show()
```

```python
#Generate weights as a KxN matrix where K is the number of eigenfaces and N the number of samples
weights = eigenfaces @ (facematrix - pca.mean_).T
```

```python
#Vamos testar primeiro em uma pessoa que já está no banco de dados

query = faces["Faces/Abdullah_Gul_0009.pgm"].reshape(1,-1)
query_weight = eigenfaces @ (query - pca.mean_).T
euclidean_distance = np.linalg.norm(weights - query_weight, axis=0)
best_match = np.argmin(euclidean_distance)
print("Best match %s with Euclidean distance %f" % (facelabel[best_match], euclidean_distance[best_match]))


fig, axes = plt.subplots(1,2,sharex=True,sharey=True,figsize=(8,6))
axes[0].imshow(query.reshape(faceshape), cmap="gray")
axes[0].set_title("Query")
axes[1].imshow(facematrix[best_match].reshape(faceshape), cmap="gray")
axes[1].set_title("Best match")
plt.show()
```

```python
#Agora em uma pessoa que não está no banco de dados

query = faces["Faces/Alvaro_Uribe_0006.pgm"].reshape(1,-1)
query_weight = eigenfaces @ (query - pca.mean_).T
euclidean_distance = np.linalg.norm(weights - query_weight, axis=0)
best_match = np.argmin(euclidean_distance)
print("Best match %s with Euclidean distance %f" % (facelabel[best_match], euclidean_distance[best_match]))


fig, axes = plt.subplots(1,2,sharex=True,sharey=True,figsize=(8,6))
axes[0].imshow(query.reshape(faceshape), cmap="gray")
axes[0].set_title("Query")
axes[1].imshow(facematrix[best_match].reshape(faceshape), cmap="gray")
axes[1].set_title("Best match")
plt.show()
```

# Conclusão

Quando o rosto procurado é de uma pessoa que está na base de dados, o algoritmo identifica corretamente a pessoa. \
Por outro lado, se o rosto buscado não está na base de dados, o "match" acontece com a pessoa mais similar, ou seja, com a menor distância euclidiana possível. 
- É possível perceber que óculos e adereços interferem no "matching", dificultando o reconhecimento facial. 

Afinal, concluimos que apesar de não ser um método super complexo, o uso de eigenfaces é suficientemente eficiente no uso de reconhecimento facial e funciona com uma ótima precisão. Além disso, fatores como a quantidade de imagens de cada pessoa no banco de dados, a qualidade dessas imagens e a iluminação também interferem no resultado, facilitando ou dificultando a identificação.
