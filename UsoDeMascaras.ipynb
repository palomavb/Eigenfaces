# Previsão do Uso de Máscaras pelo Egenfaces

```python
import cv2
import zipfile
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
faces = {}
with zipfile.ZipFile("Fotos.zip") as facezip:
    for filename in facezip.namelist():
        with facezip.open(filename) as image:
            faces[filename] = cv2.imdecode(np.frombuffer(image.read(), np.uint8), cv2.IMREAD_GRAYSCALE)
```

```python
classes = set(filename.split("_0")[0] for filename in faces.keys())
types = set (filename.split("_")[0] for filename in faces.keys())
print("Número de classes:", len(classes))
print("As classes:", (classes))
print("Os tipos:", types )
```

```python
facematrix = []
facelabel = []
for key,val in faces.items():
    if key == "Fotos/Mascara_cinza_001.jpeg":
        continue # foto teste
    if key == "Fotos/Mascara_azul_010.jpeg":
        continue # foto teste
    if key == "Fotos/Mascara_preta_010.jpeg":
        continue # foto teste
    if key == "Fotos/Mascara_branca_010.jpeg":
        continue # foto teste
    if key == "Fotos/Sem_010.jpeg":
        continue # foto teste
    facematrix.append(val.flatten())
    facelabel.append(key.split("_0")[0]) #adiciona o 0, pois caso contrário corta em máscara
 
facematrix = np.array(facematrix)
print(len(facematrix))
```

```python
pca = PCA().fit(facematrix)
print(pca.explained_variance_ratio_)
```

```python
faceshape = list(faces.values())[20].shape
print("Tamanho das imagens:", faceshape)
#todas foram cortadas utilizando o site https://www.iloveimg.com/crop-image a fim de garantir que tivessem o mesmo tamanho
```

```python
n_components = 36
eigenfaces = pca.components_[:n_components]
 
fig, axes = plt.subplots(4,5,sharex=True,sharey=True,figsize=(8,10))
for i in range(20):
    axes[i%4][i//4].imshow(eigenfaces[i].reshape(faceshape), cmap="gray")
plt.show()
```

```python
weights = eigenfaces @ (facematrix - pca.mean_).T
```

```python
#Testa uma foto de cada tipo de máscara
query1 = faces["Fotos/Mascara_cinza_001.jpeg"].reshape(1,-1)
query_weight1 = eigenfaces @ (query1 - pca.mean_).T
euclidean_distance1 = np.linalg.norm(weights - query_weight1, axis=0)
best_match1 = np.argmin(euclidean_distance1)
print("Best match %s with Euclidean distance %f" % (facelabel[best_match1], euclidean_distance1[best_match1]))
if  (facelabel[best_match1].startswith('Fotos/Mascara')):
    print("O algoritmo reconheceu a máscara!")

fig, axes = plt.subplots(1,2,sharex=True,sharey=True,figsize=(8,6))
axes[0].imshow(query1.reshape(faceshape), cmap="gray")
axes[0].set_title("Query")
axes[1].imshow(facematrix[best_match1].reshape(faceshape), cmap="gray")
axes[1].set_title("Best match")
```

# Conclusão

## Requisitos para reconhecimentos de máscara

- Precisamos de fotos com pelo menos duas tonalidades de máscaras (clara e escura) para que o algoritmo funcione melhor
- Fotos de vários ângulos da mesma pessoa, com o rosto centralizado

## Pontos positivos
- Reconhece diferentes tipos de máscara tendo pelo menos duas tonalidades no banco para análise, não sendo necessário cadastrar uma nova máscara todas as vezes
    - Máscaras azuis e cinzas foram identificadas como brancas, isto é, formam um grupo de máscaras de tonalidade "clara"
- Em máscaras de alto contraste com a pele, é possível perceber claramente sua falta
- O match quando a pessoa está sem máscara é bem maior, então é possível diferenciar claramente a utilização da máscara ou não

## Problemas encontrados
 - A maioria das fotos de câmeras de segurança em estabelecimentos comerciais não captura naturalmente os rostos das pessoas na angulação necessária
- É necessário o mesmo tamanho de pixels para todas as imagens, o que não é viável em um software de isolamento de faces em uma loja, por exemplo
- Reconhecimento em situações muito específicas, quando a máscara for retirada completamente
